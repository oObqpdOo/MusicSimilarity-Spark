```python
#!/usr/bin/python
# -*- coding: utf-8 -*-

import pandas
import matplotlib.pyplot as plt

from string import ascii_letters
import seaborn as sns
import pandas as pd
import numpy as np

sns.set(style='ticks')
```

```python
rs = np.random.RandomState(0)
data = pd.read_csv("__feature_eval.csv")[["rp","rh","bh","js","skl","mfcc","chroma","notes","agg"]]
corr = data.corr()
corr.style.background_gradient(cmap='coolwarm', low=0.5, axis=None)

```

```python
corr = data.corr()
corr.style.background_gradient(cmap='PRGn', axis=None)
# Generate a custom diverging colormap
cmap = sns.diverging_palette(220, 10, as_cmap=True)
# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11,11))
plt.rcParams.update({'font.size': 12})
plt.title('Correlation of features, 95 songs (5 per genre), 1517 artists')

sns.heatmap(corr, cmap=cmap,
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values, 
            vmax=.3, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5})
```
