```python
#!/usr/bin/python
# -*- coding: utf-8 -*-

import pandas
import matplotlib.pyplot as plt


from string import ascii_letters
import seaborn as sns
import pandas as pd
import numpy as np

rs = np.random.RandomState(0)
data = pd.read_csv("Songs_Klassik_Rachmaninoff-IdilBiret-Op3_No2PreludeinC-SharpMinor.csv")[[
"scaled_rp","scaled_rh","scaled_bh","bpm","scaled_js","scaled_skl","scaled_mfcc","scaled_corr",
"scaled_levenshtein","aggregated"]]
# Compute the correlation matrix
corr = data.corr()
corr.style.background_gradient(cmap='RdYlBu', axis=None)

```
