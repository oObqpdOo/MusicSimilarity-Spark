```python
import pandas as pd
import matplotlib.pyplot as plt

def heatMap(df, titletext):
    #Create Correlation df
    corr = df.corr()
    corr.style.background_gradient(cmap='PRGn', axis=None)
    #Plot figsize
    fig, ax = plt.subplots(figsize=(11,11))
    #Generate Color Map
    cmap = sns.diverging_palette(220, 10, as_cmap=True)
    #Generate Heat Map, allow annotations and place floats in map
    sns.set(font_scale=1.8)
    #sns.heatmap(corr, cmap=colormap, annot=True, fmt=".2f")
    sns.heatmap(corr, cmap=cmap,
            xticklabels=corr.columns.values,
            yticklabels=corr.columns.values, 
            vmax=.5, center=0,
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot_kws={"size": 12}, annot=True)

    #g.ax_heatmap.set_xticklabels(g.ax_heatmap.get_xmajorticklabels(), fontsize = 10)
    #Apply xticks
    #plt.xticks(range(len(corr.columns)), corr.columns);
    #Apply yticks
    #plt.yticks(range(len(corr.columns)), corr.columns)
    # Set up the matplotlib figure
    plt.title(titletext)
    #show plot
    plt.show()
    

#__feature_eval_1517
plt.rcParams.update({'font.size': 16})

data = pd.read_csv("combined_csv.csv")[["rp","rh","bh","js","skl","mfcc","cc","lev","weight"]]
#pd.plotting.scatter_matrix(data, alpha = 0.3, figsize = (20,20), diagonal='hist');
#plt.show()

plt.show()
plt.figure()

#__feature_eval_1517
plt.rcParams.update({'font.size': 14})

heatMap(data, 'Pearson correlation coefficient')
```
